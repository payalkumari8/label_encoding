# There is two way to use Label_encoding approach :

Approach 1: scikit-learn library

```python
from sklearn.preprocessing import LabelEncoder 
```

```python
#create an instance of labelencoder
labelencoder = LabelEncoder()

#Assign numberical values and storing it in another column called "City_N"
covid["State_N"] = labelencoder.fit_transform(covid["State/UnionTerritory"])
```

Approach 2: Category codes

```python
covid['State/UnionTerritory'] = covid['State/UnionTerritory'].astype('category')
covid.dtypes
```

```python
covid['State/UnionTerritory'] = covid['State/UnionTerritory'].cat.codes
covid
```
